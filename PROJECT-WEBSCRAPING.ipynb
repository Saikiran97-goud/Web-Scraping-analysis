```python
Productname=[]
Price=[]
Rating=[]
Features=[]
PercentageOff=[]
OriginalPrice=[]
Pagenum=[]
for i in range(1,24):
    start_time=time.time()
    URL='https://www.flipkart.com/search?q=smart+tv&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&page={}'.format(i)
    page=requests.get(URL)
    pagecontent=page.text
    soup=BeautifulSoup(pagecontent)
    
    for x in soup.findAll('div',attrs={'class','_3pLy-c row'}):
        pname=x.find('div',attrs={'class':'_4rR01T'})
        specs=x.find('div',attrs={'class','fMghEO'})
        rat=x.find('div',attrs={'class':'_3LWZlK'})
        price = x.find('div', attrs={'class':'_30jeq3 _1_WHN1'})
        off=x.find('div', attrs={'class':'_3Ay6Sb'})
        originalprice=x.find('div', attrs={'class':'_3I9_wc _27UcVY'})
        
        
        if pname is None:
            Productname.append(np.nan)
        else:
            Productname.append(pname.text)
            
        if specs is None:
            Features.append(np.NaN)
        else:
            Features.append(specs.text)
            
        if rat is None:
            Rating.append(np.NaN)
        else:
            Rating.append(rat.text)
            
        if price is None:
            Price.append(np.NaN)
        else:
            Price.append(price.text)
            
        if off is None:
            PercentageOff.append(0)
        else:
            PercentageOff.append(off.text)
            
        if originalprice is None:
            OriginalPrice.append(price.text)
        else:
            OriginalPrice.append(originalprice.text)
            
        Pagenum.append(i)
    print('Page {} completed in {} seconds'.format(i,time.time()-start_time))
                
```
